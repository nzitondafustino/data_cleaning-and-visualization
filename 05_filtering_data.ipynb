```python
import pandas as pd
```

```python
best_sellers = pd.read_csv("DataAnalysisCourseMaterials/DataAnalysis/data/bestsellers.csv")
best_sellers
```

```python
author = best_sellers["Author"] == "Pete Souza"
best_sellers[author]
```

```python
price = best_sellers["Price"] < 10
best_sellers[price]
```

```python
prices = best_sellers["Price"].between(50,60)
best_sellers[prices]
```

```python
authors = ["Kristin Hannah","Andy Weir","Delia Owens"]
best_sellers[best_sellers["Author"].isin(authors)]
```

```python
genre = best_sellers["Genre"] == "Non Fiction"
rating = best_sellers["User Rating"] == 4.9
best_sellers[genre & rating]
```

```python
genre = best_sellers["Genre"] == "Fiction"
best_sellers[genre].sort_values("Reviews").iloc[0]["Name"]
```

```python
year = best_sellers["Year"] == 2012
genre = best_sellers["Genre"] == "Fiction"
best_sellers[year & genre].sort_values("Reviews").iloc[0:5]["Name"]
```

```python
rating = best_sellers["User Rating"] < 4.5
best_sellers[rating]["Author"].value_counts().head(5).plot(kind="bar")
```
