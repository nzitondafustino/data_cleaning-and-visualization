```python
import matplotlib.pyplot as plt
```

```python
plants = ["Spinach","Turnip","Rhubarb","Broccoli","Kale"]
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="orchid")
plt.title("Germinated Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,color="red")
plt.title("died Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,color="red",width=.4)
plt.bar(plants,germinated,color="green",width=.4,align="edge")
plt.title("died Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,color="red",bottom=germinated,label="Failed seedlings")
plt.bar(plants,germinated,color="green",label="Germinated Seedlings")
plt.legend()
plt.title("died Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants,died,color="red")
plt.barh(plants,germinated,color="green",label="Germinated Seedlings")
plt.legend()
plt.title("died Rates")
plt.ylabel("Number Of Seeldings")
```

```python
titanic = pd.read_csv("DataAnalysisCourseMaterials/DataAnalysis/data/titanic.csv")
```

```python
titanic.age = pd.to_numeric(titanic["age"],errors="coerce")
```

```python
titanic.age
```

```python
plt.hist(titanic.age)
```

```python
first_class = titanic[titanic["pclass"]==1]["age"]
second_class = titanic[titanic["pclass"]==2]["age"]
third_class = titanic[titanic["pclass"]==3]["age"]
plt.hist(first_class,label="First class",alpha=.5)
plt.hist(second_class,label="second class",alpha=.5)
plt.hist(third_class,label="Third class",alpha=.5)
plt.legend()
```

```python
resorts = ["Crested Butte","Aspen","Telluride","Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure()
plt.bar(resorts,snow_last_24,label="last 24 hrs",color="#74b9ff")
plt.bar(resorts,snow_next_24,bottom=snow_last_24,label="next 24 hrs",color="#7ed6df")
plt.title("Co Snow Storm Report")
plt.ylabel("Inches")
```
