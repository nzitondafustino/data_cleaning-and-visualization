```python
import pandas as pd
houses = pd.read_csv("kc_house_data.csv",parse_dates=["date"])
```

```python
data = houses["date"].dt.dayofweek.value_counts().sort_index()
indexes = data.index

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(indexes,data)
plt.xticks(indexes,["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
plt.title("Sales By Week Day")

data = houses["date"].dt.month.value_counts().sort_index()
indexes = data.index
plt.subplot(1,2,2)
plt.plot(indexes,data)
# plt.xticks(indexes,["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
plt.title("Sales By Month")
plt.ylabel("Home sales")
plt.xlabel("Month")
plt.show()
```

```python
ufos = pd.read_csv("nuforc_reports.csv",parse_dates=["date_time"])
```

```python
ufos[ufos["state"] == "CA"]["date_time"].dt.year.value_counts().sort_index().plot(label="CA")
ufos[ufos["state"] == "TX"]["date_time"].dt.year.value_counts().sort_index().plot(label="TX")
plt.legend()
```

```python
ufos[ufos["shape"]=="light"].date_time.dt.year.value_counts().sort_index().plot(
    title="UFOS Sighting by shape",
    figsize=(10,8)
)
ufos[ufos["shape"]=="fireball"].date_time.dt.year.value_counts().sort_index().plot()
ufos[ufos["shape"]=="circle"].date_time.dt.year.value_counts().sort_index().plot()
ufos[ufos["shape"]=="triangle"].date_time.dt.year.value_counts().sort_index().plot()
ufos[ufos["shape"]=="formation"].date_time.dt.year.value_counts().sort_index().plot()
plt.legend(["Light","Fireball","Cicle","Triangle","Formation"])
plt.xlim(2000,2018)
plt.show()
```
